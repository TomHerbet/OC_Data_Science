```python
import openfoodfacts
import csv

api = openfoodfacts.API(user_agent="MonAppDataScience/1.0 (thomas.herbet@gmail.com)")
res = api.product.text_search("champagne")

products = res.get("products", [])[:10]

with open("champagne_off_sdk.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["foodId","label","category","foodContentsLabel","image"])
    writer.writeheader()
    for prod in products:
        writer.writerow({
            "foodId": prod.get("code"),
            "label": prod.get("product_name"),
            "category": prod.get("brands"),
            "foodContentsLabel": prod.get("ingredients_text"),
            "image": prod.get("image_front_url"),
        })

print("SDK CSV généré.")
```
